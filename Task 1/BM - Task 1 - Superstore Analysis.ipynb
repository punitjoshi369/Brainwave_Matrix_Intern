### <span style = 'color: red'> Columns: Order_Date and Ship_Date requires necessary "datatype" change.</span>

```python
sales['Order_Date'] = pd.to_datetime(sales['Order_Date'], format = "%d/%m/%Y")
sales['Ship_Date'] = pd.to_datetime(sales['Ship_Date'], format = "%d/%m/%Y")
```

### <span style = 'color: blue'> A quick summary check to ensure data type change is successful </span>

```python
sales.info()
```

### <span style = 'color: red'> Checking Null Values </span>

```python
sales.isnull().sum()
```

### <span style = 'color: red'> Percentage calculation of Null value </span>

```python
sales.isnull().sum()/ len(sales) * 100
```

### <span style = 'color: red'> Since null values (Postal Code) are less in number, we chose to drop those rows.</span>

```python
sales.dropna(subset = ['Postal_Code'], inplace = True)
```

### <span style = 'color: blue'> A quick Percentage calculation of Null value to ensure drop is successful </span>

```python
sales.isnull().sum()/ len(sales) * 100
```

### <span style = 'color: red'> Checking if there are any duplicated values </span>

```python
duplicate = sales[sales.iloc[:,1:].duplicated(keep=False)]
duplicate
```

### <span style = 'color: red'> Dropping the Duplicate records </span>

```python
sales = sales.iloc[:,1:].drop_duplicates()
```

### <span style = 'color: red'> Shape Check </span>

```python
sales.shape
```

### <span style = 'color: red'> Data Type check </span>

```python
sales.dtypes
```

### <span style = 'color: red'> Postal Code cannot be floating. We should have changed it earlier </span>

```python
sales['Postal_Code'] = sales['Postal_Code'].astype('int64')
```

```python
sales.dtypes
```
